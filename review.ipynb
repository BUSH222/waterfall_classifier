```python
# do not run this, you will spend too much time reviewing images

import os
import glob
import shutil
import cv2

WITH_DIR = "data/with_signal"
WITHOUT_DIR = "data/without_signal"
os.makedirs(WITHOUT_DIR, exist_ok=True)

def review_with_signal_images():
    files = sorted(glob.glob(os.path.join(WITH_DIR, "*.png")))
    print(f"Found {len(files)} images in {WITH_DIR}")

    for i, path in enumerate(files, 1):
        img = cv2.imread(path)
        if img is None:
            print(f"Could not open {path}, skipping.")
            continue

        title = f"[{i}/{len(files)}] {os.path.basename(path)}  (SPACE = keep, X = move to without_signal, Q = quit)"
        cv2.imshow("Review", img)
        cv2.setWindowTitle("Review", title)

        while True:
            key = cv2.waitKey(0) & 0xFF
            if key == ord('x'):
                dest = os.path.join(WITHOUT_DIR, os.path.basename(path))
                print(f"Moving to without_signal: {path} -> {dest}")
                shutil.move(path, dest)
                break
            elif key == 32:
                print(f"Keeping: {path}")
                break
            elif key == ord('q'):
                print("Quitting review.")
                cv2.destroyAllWindows()
                return
            else:
                print("Press SPACE to keep, X to move, Q to quit.")

    cv2.destroyAllWindows()
    print("Review finished.")

review_with_signal_images()
```
